```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
df =pd.read_csv("../data/NIFTY 50 and bank nifty -13-05-2023-to-13-05-2024.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
trainx_x_col=df.columns[1:13]
trainx_x_col
```

```python
trainx_y_col=df.columns[20:24]
trainx_y_col
```

```python
x_train, x_test, y_train, y_test = train_test_split(df[trainx_x_col], df[trainx_y_col], test_size= 0.2, random_state= 1)
```

```python
x_train
```

```python
y_train
```

```python
l1=LinearRegression()
l1.fit(x_train,y_train)
```

```python
score1=l1.score(x_test,y_test)
```

```python
score1
```
