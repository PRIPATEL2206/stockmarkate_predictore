```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import LSTM,Dense
```

```python
model=Sequential()
```

```python
def set_to_priodic(data , duration=10):
    sequencial_x_data=[]
    for i in range(len(data)-duration):
        sequencial_x_data.append(data[i:i+duration].to_numpy().tolist())
    return sequencial_x_data
```

```python
x_train=x_train.to_numpy()
```

```python
sequencial_x_data = set_to_priodic(x_train)
```

```python
model.add(LSTM(100, activation='relu', input_shape=(10, 5)))
model.add(Dense(1,activation="softmax"))
```

```python
model.compile(
    loss="mse",
    optimizer="adam",
    metrics=["accuracy"]
)
model.summary()
```

```python
len(sequencial_x_data[0][0])
```

```python
sequencial_x_data
```

```python
sequencial_y_data=y_train[10:]
```

```python
model.fit(np.asarray(sequencial_x_data),sequencial_y_data.to_numpy(),epochs=100)
```
