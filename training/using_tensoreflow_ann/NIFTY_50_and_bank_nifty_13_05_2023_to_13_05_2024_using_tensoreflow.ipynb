```python
import  tensorflow as tf
```

```python
import pandas as pd
```

```python
df=pd.read_csv("../../data/NIFTY 50 and bank nifty -13-05-2023-to-13-05-2024.csv")
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df=df.dropna()
```

```python
df.isna().sum()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
m1=tf.keras.Sequential(
    [
        tf.keras.layers.Dense(8, activation='relu'),
        tf.keras.layers.Dense(6, activation='relu'),
        tf.keras.layers.Dense(6, activation='relu'),
        tf.keras.layers.Dense(4, activation='sigmoid')
    ]
)
m1.build(input_shape=12)
```

```python
m1.summary()
```

```python
m1.compile(
    optimizer='adam',
    loss=tf.keras.losses.MeanAbsoluteError(
    reduction='sum_over_batch_size',
    name='mean_absolute_error'
),
    metrics=['accuracy']    
)
```

```python
df.columns
```

```python
trainx_y_col=df.columns[20:24]
trainx_y_col
trainx_x_col=df.columns[1:13]
trainx_x_col
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(df[trainx_x_col], df[trainx_y_col], test_size= 0.2, random_state= 1)
```

```python
x_train
```

```python
m1.fit(
    scaler.fit_transform(x_train),
   scaler.fit_transform(y_train),
    epochs=500
)
```

```python
m1.evaluate(scaler.fit_transform(x_test),scaler.fit_transform(y_test))
```
