```python
import pandas as pd
```

```python
df=pd.read_csv("../data/NIFTY 50 - Minute data1.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
# Splitting the data into train and test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df.drop(columns=["date","close","volume"]), df["close"], test_size= 0.3, random_state= 1)
```

```python
x_train.head()
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression

l1=LinearRegression()
```

```python
l1.fit(x_train,y_train)
```

```python
score=l1.score(x_test,y_test)
```

```python
score
```
